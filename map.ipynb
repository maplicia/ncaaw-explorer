```python
import folium
import pandas as pd
import geopandas as gpd
```

```python
csv = r'warrennolan_wbb2026_tournament.csv'
df = pd.read_csv(csv)
```

```python
geometry = gpd.points_from_xy(df.stadium_lng, df.stadium_lat)
geo_df = gpd.GeoDataFrame(
    df, geometry=geometry
)

geo_df.head()
```

```python
url = "https://a.espncdn.com/i/teamlogos/ncaa/500/{}.png".format

map = folium.Map(location=[39.8, -98.5], tiles="CartoDB Positron", zoom_start=4)
for i in range(0,len(geo_df)):
   icon_image = url(geo_df.iloc[i]['espn_id'])
   folium.Marker(
      location=[geo_df.iloc[i]['stadium_lat'], geo_df.iloc[i]['stadium_lng']],
      popup=geo_df.iloc[i]['team'],
      icon = folium.CustomIcon(
      icon_image,
      icon_size=(20,20))

   ).add_to(map)
   
map
```
