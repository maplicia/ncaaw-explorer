```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import plotly.express as px
import geopandas as gpd
import folium
```

```python
csv = r'/Users/alicia/Desktop/misc/ncaaw_explorer/warrennolan_wbb2026_tournament.csv'
df = pd.read_csv(csv)
```

```python
x = df.pace
y = df.net_rating
```

```python
x_axis = x.median()
y_axis = y.median()
```

```python
fig = px.scatter(df, x="pace", y="net_rating", hover_data=["team"])
fig.update_xaxes(tick0="x_axis",dtick=1)
fig.update_yaxes(tick0="y_axis",dtick=5)

fig.add_shape(
type="line", x0=61, x1=82, y0=17, y1=17, line=dict(color="black", width=1))
fig.add_shape(
type="line", x0=72, x1=72, y0=-15, y1=55, line=dict(color="black", width=1))

fig.add_shape(
type="line", x0=77, x1=67, y0=52, y1=52, line=dict(color="green", width=2))
fig.add_shape(
type="line", x0=74.5, x1=69.5, y0=27, y1=27, line=dict(color="green", width=2))
fig.add_shape(
type="line", x0=69.5, x1=67, y0=27, y1=52, line=dict(color="green", width=2))
fig.add_shape(
type="line", x0=74.5, x1=77, y0=27, y1=52, line=dict(color="green", width=2))

''' 
trapezoid points: ((69.5, 27), (67,52), (77, 52), (74.5, 27)) 

upper limits: +/- 5 pace, max net
lower limits: +/- 2.5 pace, +10 net
'''


fig.write_html('TOE.html')
```

```python
geometry = gpd.points_from_xy(df.stadium_lng, df.stadium_lat)
geo_df = gpd.GeoDataFrame(
    df, geometry=geometry
)

geo_df.head()
```

```python
map = folium.Map(location=[39.8, -98.5], tiles="CartoDB Positron", zoom_start=4)
for i in range(0,len(geo_df)):
   folium.Marker(
      location=[geo_df.iloc[i]['stadium_lat'], geo_df.iloc[i]['stadium_lng']],
      popup=geo_df.iloc[i]['team'],
   ).add_to(map)
   
map
```
